## GSD word2pos

```python
path_to_data = '/home/ubuntu/masters/data/UD_Russian-GSD/ru_gsd-ud-train.conllu'
train = pyconll.load_from_file(path_to_data)
```

```python
word2pos = {}
for i, sentence in tqdm_notebook(enumerate(train)):
    for token in sentence:
        word = token.form.lower()
        pos = token.upos
        if word not in word2pos:
            word2pos[word] = [pos]
        else:
            word2pos[word].append(pos)
```

## OpenCorpora word2pos

```python
path_to_data = '/home/ubuntu/masters/data/opencorpora/annot.opcorpora.no_ambig.xml'
corpus = opencorpora.load(path_to_data)
```

```python
opencorp_word2pos = {}
for doc in tqdm_notebook(corpus):
    if len(doc.sentences) != 0:
        for sentence in doc.sentences:
            for token in sentence:
                word = token.source.lower()
                pos = token.grammemes[0]
                if word not in opencorp_word2pos:
                    opencorp_word2pos[word] = [pos]
                else:
                    opencorp_word2pos[word].append(pos)
```

```python
for word in opencorp_word2pos:
    opencorp_word2pos[word] = set(opencorp_word2pos[word])
```

## SynTagRus word2pos

```python
path_to_corpus = '/home/ubuntu/masters/data/syntagrus/SynTagRus2018/'
```

```python
paths_to_files_syntagrus = []
for path, subdirs, files in os.walk(path_to_corpus):
    for name in files:
        paths_to_files_syntagrus.append(os.path.join(path, name))
```

```python
syntagrus_word2pos = {}
for file in tqdm_notebook(paths_to_files_syntagrus):
    file_lines = []
    with open(file, 'r') as f:
        for line in f:
            file_lines.append(line)
    for line in file_lines:
        if line.startswith('<W'):
            word = line.split('>')[1].split('<')[0].lower()
            pos = line.split(' ')[2].split('FEAT="')[-1]
            if pos.endswith('"'):
                pos = pos[:-1]
            if word not in syntagrus_word2pos:
                syntagrus_word2pos[word] = [pos]
            else:
                syntagrus_word2pos[word].append(pos)
```

```python
for word in syntagrus_word2pos:
    syntagrus_word2pos[word] = set(syntagrus_word2pos[word])
```

```python
# Change dict to defaultdict
```

```python
bundled2num = {}
num_pairs = 0
not_paired = 0
for gsd_word in word2pos:
    if gsd_word in syntagrus_word2pos:
        num_pairs += 1
        for gsd_pos in word2pos[gsd_word]:
            for syntagrus_pos in syntagrus_word2pos[gsd_word]:
                if (gsd_pos, syntagrus_pos) not in bundled2num:
                    bundled2num[(gsd_pos, syntagrus_pos)] = 1
                else:
                    bundled2num[(gsd_pos, syntagrus_pos)] += 1
        print(gsd_word, word2pos[gsd_word], syntagrus_word2pos[gsd_word])
    else:
        not_paired += 1
```
